# 2. Preprocess

Right now in my data, I have sentiment column as the target, text as the textual information to be tokenized. Features with 'tweet_' prefix are all categorical and features with 'meta_' prefix are all numerical.

### 2.1 Implement the clean and tokeinze functions

Implement my idea for texual features in the function below. I kept the option of removing stop words and stemming words, by default I'm not doing any operation as of now. I'll check later on which option optimize the model.

```python
def clean_text(text, stop=None, stem=False):
    """
    Pre process the textual tweet, return cleaned text
    """
    if type(text)!=str:
        return text
    
    # convert emoji to string
    text=emoji.demojize(text)
    # replace xml escaping characters
    text=html.unescape(text)
    # remove @ mentions
    text=re.sub(r'@[A-Za-z0-9]+','',text)
    # remove url
    text=re.sub(r'https?:\/\/[^\r\n\s]*', '', text, flags=re.MULTILINE)
    # replace puncuations to space
    text=re.sub('[%s]'%string.punctuation, ' ', text)
    # extract only letter
    letters_only=re.sub(r"[^a-zA-Z]", " ",text)
    # trasform to lower cases and separate by space
    words=letters_only.lower().split()
    
    # remove stop words
    if stop:
        words=[w for w in words if not w in stop]
    # stem the words
    if stem:
        stemmer=PorterStemmer()
        words=[stemmer.stem(w) for w in words]
    return " ".join(words)
```

Tokenize the entire dataset and count the uniq words for the given dataframe using the following function

```python
def tokenize(df, stop=None, stem=False):
    df=df.copy(deep=True)
    df['text_clean']=df.apply(lambda x: clean_text(x['text'], stop=stop, stem=stem), axis=1)
    corpus=[]
    df.apply(lambda x: corpus.extend(x['text_clean'].split()), axis=1)
    cnt=Counter(corpus)
    return df, corpus, cnt
```

```python
tweets, corpus, cnt=tokenize(tweets)
```

The generally defined stopwords contain passive words like 'no' and it might influence the model performance if I delete them directly. So I also customized stop words, which contains top 50 most frequent words from above (that are also in general words) and words having single character only. 

```python
general_stop=set(stopwords.words("english"))
len(general_stop)
```

```python
customized_stop=set([i[0] for i in cnt.most_common(50) 
                     if i[0] in general_stop]) | set([i for i in cnt if len(i)==1])
len(customized_stop)
```

```python
def stop_words(tweets):
    _, _, cnt=tokenize(tweets)
    general_stop=set(stopwords.words("english"))
    customized_stop=set([i[0] for i in cnt.most_common(50) 
                     if i[0] in general_stop]) | set([i for i in cnt if len(i)==1])
    return general_stop, customized_stop
```

```python
tokens['cnt']=tokens['token'].map(lambda x: cnt[x])
tokens['rank']=tokens['token'].map(lambda x: ranking[x])
```

```python
for l in ['negative', 'neutral', 'positive']:
    _, _, c=tokenize(tweets[tweets['sentiment']==l])
    tokens[l]=tokens['token'].map(lambda x: 0 if x not in c else c[x])
    tokens[l+'_ratio']=tokens[l]/tokens['cnt']
    tokens[l+'_pct']=tokens[l]/tokens[l].sum()
```

I created a dataframe for all the tokens. _ratio columns are the class count over the total count of the token. _pct columns are the class count over the total count of the class.

### 2.3 Handle categorical features

```python
def one_hot(df, columns, fillna=None):
    if type(columns)!=list:
        columns=[columns]
        
    df=df.copy(deep=True)
    if fillna:
        df=df.apply(fillna, 1)
    else:
        df=df.fillna('')
        
    for c in columns:
        df=df.join(pd.get_dummies(df.pop(c), prefix=c))
    
    return df
```
